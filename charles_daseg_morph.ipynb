```python
from morphometrics import get_morphometrics
import pandas as pd
import os
```

```python
dir = r"D:\Research\Isaacs Lab\DeepAxon\Charles Image DA Seg\Segmented"
```

```python
for img in os.listdir(dir):
    img_path = os.path.join(dir, img)
    img_name = img.split('.')[0]
    df = get_morphometrics(img_path)
    df = df[df['myelin_thickness']>0]
    df=  df[df['gratio']>0]
    print(img_name + '\t' + str(len(df)) + '\t' + str(df['gratio'].mean()) + '\t' + str(df['axon_diam'].mean()*0.106))
```
