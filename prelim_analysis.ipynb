```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
df = pd.read_excel(r"D:\Research\Isaacs Lab\DeepAxon\Charles Data.xlsx", sheet_name='Full Data (Per Image)')
```

```python
ads = df[df['software']=='ads']
ip = df[df['software']=='ip']
halo = df[df['software']=='halo']
da_major = df[df['software']=='da_major']
da_avg = df[df['software']=='da_avg']
```

```python
plt.hist(ads['axon_count'], alpha=0.5, label='ads')
plt.hist(ip['axon_count'], alpha=0.5, label='ip')
plt.hist(halo['axon_count'], alpha=0.5, label='halo')
plt.hist(da_major['axon_count'], alpha=0.5, label='da')
# plt.hist(da_avg['axon_count'], alpha=0.5, label='da_avg')

plt.title('Axon Counts')
plt.xlabel('Axon Count')
plt.ylabel('Freq')
plt.legend()
plt.show()
```

```python
# plt.hist(ads['gratio_avg'], alpha=0.5, label='ads')
# plt.hist(ip['gratio_avg'], alpha=0.5, label='ip')
# plt.hist(halo['gratio_avg'], alpha=0.5, label='halo')
plt.hist(da_major['gratio_avg'], alpha=0.5, label='da_major')
plt.hist(da_avg['gratio_avg'], alpha=0.5, label='da_avg')

plt.title('Average G-ratio')
plt.xlabel('G-ratio')
plt.ylabel('Freq')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.hist(ads['diamter_avg'], alpha=0.5, label='ads')
plt.hist(halo['diamter_avg'], alpha=0.5, label='halo')
plt.hist(da_major['diamter_avg'], alpha=0.5, label='da_major')
plt.hist(da_avg['diamter_avg'], alpha=0.5, label='da_avg')

plt.title('Average Axon Diameter')
plt.xlabel('Diameter')
plt.ylabel('Freq')
plt.legend()
plt.tight_layout()
plt.show()
```
